Data Modelling and Evaluation

```python
X=dr.drop('INGREDIENT_NAME', axis=1)
y=dr['INGREDIENT_NAME']
X=X.drop('POTENCY_AMOUNT',axis=1)
```

```python
X=pd.get_dummies(X, columns=['ROUTE', 'DOSAGE_FORM', 'UNII'])
X
```

```python
y=pd.get_dummies(y, columns=['INGREDIENT_NAME'])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
multi_output_classifier = MultiOutputClassifier(DecisionTreeClassifier(random_state=0))
multi_output_classifier.fit(X_train, y_train)
```

```python
y_pred = multi_output_classifier.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
```

```python
from sklearn.metrics import accuracy_score, hamming_loss, f1_score, roc_auc_score, average_precision_score
import numpy as np
```

```python
hamming = hamming_loss(y_test, y_pred)
print("Hamming Loss:", hamming)

# calculate F1 score
f1 = f1_score(y_test, y_pred, average='macro')  # can also use 'micro' or 'weighted' average depending on the problem
print("F1 Score:", f1)
```

```python
# calculate AUC-PR score
pr_auc = average_precision_score(y_test, y_pred, average='macro')  # can also use 'micro' or 'weighted' average depending on the problem
print("AUC-PR Score:", pr_auc)
```

```python
from sklearn.inspection import permutation_importance

# calculate permutation feature importance
result = permutation_importance(multi_output_classifier, X_test, y_test, n_repeats=10, random_state=0)

# plot the feature importance
fig, ax = plt.subplots(figsize=(10,6))
sorted_idx = result.importances_mean.argsort()
ax.barh(X_test.columns[sorted_idx], result.importances_mean[sorted_idx])
ax.set_title("Permutation Feature Importance")
plt.show()
```
